```python
def extract_frames(video_path, frame_interval=5):
    """
    Extract frames from a video at every 'frame_interval' seconds.
    """
    frames = []
    video = cv2.VideoCapture(video_path)
    if not video.isOpened():
        print(f"Error opening video file: {video_path}")
        return frames

    fps = video.get(cv2.CAP_PROP_FPS)
    if fps == 0 or fps is None or np.isnan(fps):
        fps = 25  # Default FPS if unable to get
    total_frames = int(video.get(cv2.CAP_PROP_FRAME_COUNT))
    duration = total_frames / fps
    frame_numbers = [int(fps * sec) for sec in range(0, int(duration), frame_interval)]

    for frame_num in frame_numbers:
        video.set(cv2.CAP_PROP_POS_FRAMES, frame_num)
        success, frame = video.read()
        if success:
            frames.append(frame)
        else:
            break
    video.release()
    return frames
```

```python
def detect_faces_and_get_embeddings(frame):
    """
    Detect faces in a frame using InsightFace's FaceAnalysis.
    Returns a list of embeddings and bounding boxes.
    """
    try:
        faces = app.get(frame)
        embeddings = []
        face_boxes = []
        for face in faces:
            embedding = face.embedding
            embeddings.append(embedding.tolist())  # Convert numpy array to list
            bbox = face.bbox.astype(int)
            x1, y1, x2, y2 = bbox
            face_boxes.append((x1, y1, x2 - x1, y2 - y1))  # (x, y, w, h)
        return embeddings, face_boxes
    except Exception as e:
        print(f"Error in detect_faces_and_get_embeddings: {e}")
        return [], []
```

```python
from scipy.spatial.distance import cosine

class SimpleFaceTracker:
    def __init__(self, threshold=0.5, max_lost=5):
        """
        Initializes the tracker.

        Parameters:
        - threshold: Cosine similarity threshold for matching faces.
        - max_lost: Maximum number of frames to keep 'lost' tracks before removing them.
        """
        self.threshold = threshold
        self.max_lost = max_lost
        self.next_id = 1
        self.tracks = {}  # track_id: {'embedding': ..., 'bbox': ..., 'lost': 0}

    def update(self, embeddings, bboxes):
        """
        Update tracker with new detections.

        Parameters:
        - embeddings: List of face embeddings.
        - bboxes: List of bounding boxes (x, y, w, h).

        Returns:
        - List of track IDs corresponding to the current detections.
        """
        updated_ids = []
        unmatched_tracks = list(self.tracks.keys())
        unmatched_detections = list(range(len(embeddings)))

        if len(self.tracks) == 0:

            for det in range(len(embeddings)):
                self.tracks[self.next_id] = {
                    'embedding': embeddings[det],
                    'bbox': bboxes[det],
                    'lost': 0
                }
                updated_ids.append(self.next_id)
                self.next_id += 1
            return updated_ids

         track_ids = list(self.tracks.keys())
        similarities = np.zeros((len(track_ids), len(embeddings)))
        for t, track_id in enumerate(track_ids):
            for d, emb in enumerate(embeddings):
                similarities[t, d] = 1 - cosine(self.tracks[track_id]['embedding'], emb)


        while True:
            if similarities.size == 0:
                break
            t, d = np.unravel_index(similarities.argmax(), similarities.shape)
            max_sim = similarities[t, d]
            if max_sim < self.threshold:
                break
            track_id = track_ids[t]
            updated_ids.append(track_id)

            self.tracks[track_id]['embedding'] = embeddings[d]
            self.tracks[track_id]['bbox'] = bboxes[d]
            self.tracks[track_id]['lost'] = 0
            # Remove matched track and detection
            similarities = np.delete(similarities, t, axis=0)
            similarities = np.delete(similarities, d, axis=1)
            track_ids.pop(t)
            unmatched_tracks.remove(track_id)
            unmatched_detections.remove(d)


        for track_id in unmatched_tracks:
            self.tracks[track_id]['lost'] += 1
            if self.tracks[track_id]['lost'] > self.max_lost:
                del self.tracks[track_id]


        for det in unmatched_detections:
            self.tracks[self.next_id] = {
                'embedding': embeddings[det],
                'bbox': bboxes[det],
                'lost': 0
            }
            updated_ids.append(self.next_id)
            self.next_id += 1

        return updated_ids
```

```python
from scipy.spatial.distance import cosine
from scipy.optimize import linear_sum_assignment

class SimpleFaceTracker:
    def __init__(self, threshold=0.5, max_lost=5):
        """
        Initializes the tracker.

        Parameters:
        - threshold: Cosine similarity threshold for matching faces.
        - max_lost: Maximum number of frames to keep 'lost' tracks before removing them.
        """
        self.threshold = threshold
        self.max_lost = max_lost
        self.next_id = 1
        self.tracks = {}  # track_id: {'embedding': ..., 'bbox': ..., 'lost': 0}

    def update(self, embeddings, bboxes):
        """
        Update tracker with new detections.

        Parameters:
        - embeddings: List of face embeddings.
        - bboxes: List of corresponding bounding boxes (x, y, w, h).

        Returns:
        - List of track IDs corresponding to the current detections.
        """
        updated_ids = []
        unmatched_tracks = set(self.tracks.keys())
        unmatched_detections = set(range(len(embeddings)))

        if len(self.tracks) == 0:
            # Initialize new tracks
            for det in range(len(embeddings)):
                self.tracks[self.next_id] = {
                    'embedding': embeddings[det],
                    'bbox': bboxes[det],
                    'lost': 0
                }
                updated_ids.append(self.next_id)
                self.next_id += 1
            return updated_ids

        # Compute similarity matrix (cosine similarity)
        track_ids = list(self.tracks.keys())
        num_tracks = len(track_ids)
        num_detections = len(embeddings)

        similarity_matrix = np.zeros((num_tracks, num_detections))
        for t, track_id in enumerate(track_ids):
            for d, emb in enumerate(embeddings):
                similarity_matrix[t, d] = 1 - cosine(self.tracks[track_id]['embedding'], emb)

        # Apply the Hungarian Algorithm for optimal matching
        row_ind, col_ind = linear_sum_assignment(-similarity_matrix)  # maximize similarity

        # Keep track of matches and unmatched tracks/detections
        for t, d in zip(row_ind, col_ind):
            sim_score = similarity_matrix[t, d]
            if sim_score >= self.threshold:
                track_id = track_ids[t]
                updated_ids.append(track_id)
                # Update track
                self.tracks[track_id]['embedding'] = embeddings[d]
                self.tracks[track_id]['bbox'] = bboxes[d]
                self.tracks[track_id]['lost'] = 0
                # Remove from unmatched sets
                unmatched_tracks.discard(track_id)
                unmatched_detections.discard(d)

        # Increment lost count for unmatched tracks
        for track_id in list(unmatched_tracks):
            self.tracks[track_id]['lost'] += 1
            if self.tracks[track_id]['lost'] > self.max_lost:
                del self.tracks[track_id]

        # Initialize new tracks for unmatched detections
        for det in unmatched_detections:
            self.tracks[self.next_id] = {
                'embedding': embeddings[det],
                'bbox': bboxes[det],
                'lost': 0
            }
            updated_ids.append(self.next_id)
            self.next_id += 1

        return updated_ids
```

```python
from scipy.spatial.distance import cosine
import numpy as np

def cosine_similarity(emb1, emb2):
    """
    Calculate the cosine similarity between two embeddings.
    Returns a value between -1 and 1, where 1 means identical.
    """
    return 1 - cosine(emb1, emb2)
```

```python
import cv2
import os
import matplotlib.pyplot as plt

def save_representative_image(influencer_id, target_embedding, influencers_dir, video_filename, df, video_influencers, all_embeddings, threshold=0.5):
    """
    Save a representative image for each influencer by finding the frame where the embedding matches.

    Parameters:
    - influencer_id (int): Unique ID of the influencer.
    - target_embedding (list or np.ndarray): The embedding to match.
    - influencers_dir (str): Directory path to save the images.
    - video_filename (str): Name of the video file to search.
    - df (pd.DataFrame): DataFrame containing video data.
    - video_influencers (list): List mapping videos to influencer IDs.
    - all_embeddings (list): List of all embeddings.
    - threshold (float): Cosine similarity threshold for matching.

    Returns:
    - None
    """
    video_path = os.path.join(VIDEO_DIR, video_filename)

    if not os.path.exists(video_path):
        print(f"Video file '{video_filename}' does not exist. Skipping influencer {influencer_id}.")
        return

    frames = extract_frames(video_path, frame_interval=5)  # Extract frames every 5 seconds

    # Iterate through frames to find a matching embedding
    for frame_idx, frame in enumerate(frames):
        embeddings, face_boxes = detect_faces_and_get_embeddings(frame)

        if not face_boxes:
            continue  # No faces detected in this frame

        for emb, bbox in zip(embeddings, face_boxes):
            similarity = cosine_similarity(target_embedding, emb)
            if similarity >= threshold:
                # Valid match found
                x, y, w, h = bbox
                x, y, w, h = int(x), int(y), int(w), int(h)
                frame_height, frame_width = frame.shape[:2]

                # Validate bounding box coordinates
                if x < 0 or y < 0 or (x + w) > frame_width or (y + h) > frame_height:
                    print(f"Invalid bounding box for Influencer {influencer_id} in Frame {frame_idx + 1}. Skipping this face.")
                    continue

                face = frame[y:y+h, x:x+w]

                # Check if the face region is not empty
                if face.size == 0:
                    print(f"Empty face region for Influencer {influencer_id} in Frame {frame_idx + 1}. Skipping.")
                    continue

                # Optional: Resize the face image for consistency
                face_resized = cv2.resize(face, (224, 224))  # Example size

                # Save the face image
                image_path = os.path.join(influencers_dir, f'influencer_{influencer_id}.jpg')
                success = cv2.imwrite(image_path, face_resized)

                if success:
                    print(f"Representative image for Influencer {influencer_id} saved successfully at '{image_path}'.")
                else:
                    print(f"Failed to save image for Influencer {influencer_id} at '{image_path}'.")

                return  # Save only one image per influencer

    # If no matching embedding found, attempt to save the first detected face
    if len(frames) > 0:
        for frame_idx, frame in enumerate(frames):
            embeddings, face_boxes = detect_faces_and_get_embeddings(frame)

            if not face_boxes:
                continue

            # Save the first detected face as the representative image
            emb = embeddings[0]
            bbox = face_boxes[0]
            x, y, w, h = bbox
            x, y, w, h = int(x), int(y), int(w), int(h)
            frame_height, frame_width = frame.shape[:2]

            # Validate bounding box coordinates
            if x < 0 or y < 0 or (x + w) > frame_width or (y + h) > frame_height:
                print(f"Invalid bounding box for Influencer {influencer_id} in Frame {frame_idx + 1}. Skipping this face.")
                continue

            face = frame[y:y+h, x:x+w]

            if face.size == 0:
                print(f"Empty face region for Influencer {influencer_id} in Frame {frame_idx + 1}. Skipping.")
                continue

            face_resized = cv2.resize(face, (224, 224))  # Example size

            # Save the face image
            image_path = os.path.join(influencers_dir, f'influencer_{influencer_id}.jpg')
            success = cv2.imwrite(image_path, face_resized)

            if success:
                print(f"No matching embedding found. Representative image for Influencer {influencer_id} saved at '{image_path}'.")
            else:
                print(f"Failed to save image for Influencer {influencer_id} at '{image_path}'.")

            return  # Save only one image per influencer

    print(f"No representative image found for Influencer {influencer_id}.")
```

```python
from IPython.display import HTML
import base64

def encode_image_to_base64(image_path):
    """
    Encode an image to base64 for embedding in HTML.
    """
    if not os.path.exists(image_path):
        return 'No Image'
    with open(image_path, "rb") as img_file:
        encoded = base64.b64encode(img_file.read()).decode()
    return f'<img src="data:image/jpeg;base64,{encoded}" width="100">'

# Add image paths to the DataFrame
influencer_performance['image_path'] = influencer_performance['influencer_id'].apply(
    lambda x: os.path.join(INFLUENCERS_DIR, f'influencer_{x}.jpg')
)

# Encode images to base64
influencer_performance['image_base64'] = influencer_performance['image_path'].apply(encode_image_to_base64)

# Create HTML table with styling
html = """
<style>
table {
  width: 100%;
  border-collapse: collapse;
}
th, td {
  border: 1px solid #dddddd;
  text-align: center;
  padding: 8px;
}
th {
  background-color: #f2f2f2;
}
</style>
<table>
    <tr>
        <th>Influencer</th>
        <th>Average Performance</th>
        <th>Videos Count</th>
    </tr>
"""

for _, row in influencer_performance.iterrows():
    influencer_img = row['image_base64']
    avg_perf = row['average_performance']
    video_count = row['video_count']
    html += f"""
    <tr>
        <td>{influencer_img}</td>
        <td>{avg_perf:.2f}</td>
        <td>{video_count}</td>
    </tr>
    """

html += "</table>"

# Display the table
display(HTML(html))
```
